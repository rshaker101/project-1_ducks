```python
# Anneth's graph


df = pd.read_csv(r'C:\Users\rania\Downloads\top_insta_influencers_data.csv')


df['followers'] = df['followers'].str.replace('m', '')


df['followers'] = pd.to_numeric(df['followers'], errors='coerce')


df.sort_values(by=['followers'], ascending=False, inplace=True)


top_10_influencers = df.iloc[:10, :]


input_names = top_10_influencers['channel_info'].to_list()


top_10_influencers.loc[:, 'input_name'] = input_names


plt.figure(figsize=(10, 6))
sns.barplot(
    x="input_name",
    y="followers",
    data=top_10_influencers,
    palette="bright",
)


plt.xticks(rotation=45)


plt.title("Top 10 Influencers by Number of Followers")
plt.xlabel("Influencer Name")
plt.ylabel("Number of Followers (Millions)")

plt.show()
```

```python
df['posts']=df['posts'].apply(lambda x: int(x) if "k" not in x else int(float(x.replace('k',""))*1000) )
posts_count_on_country=df.groupby(['country'],as_index=False)['posts'].sum()
posts_count_on_country.reset_index(drop=True,inplace=True)
posts_count_on_country.sort_values(['posts'],inplace=True,ascending=False)
plt.figure(figsize=(16, 8))
plt.title("Bar chart showing total numbers of posts across countries")
ax=sns.barplot(y='country', x='posts', data=posts_count_on_country);
plt.show()
```

```python
def mea(df,col):#to replace null values with median
    
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    IQR=q3-q1
    upper=q3+(1.5*IQR)
    lower=q1-(1.5*IQR)
    median=df[col].median()
    df[col]=np.where(df[col]>upper,median,np.where(df[col]<lower,median,df[col]))
    sns.boxplot(df[col],data=df)
pd.set_option('display.max_rows', 100)
df.head(100)
```
